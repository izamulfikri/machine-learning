## Tugas 1: Mushroom Dataset with Decision Tree & Random Forest Algorithm

1. Import Library

```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier # import DT
from sklearn.ensemble import RandomForestClassifier # import RandomForest
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
import math
```

2. Load Dataset

```python
# Load data
df = pd.read_csv('data/mushrooms.csv')

df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
# Cek data imbalance

class_counts = df['class'].value_counts()

print(class_counts)


# Membagi fitur
x = df.loc[:,'cap-shape':'habitat']
y = df['class']

x = pd.get_dummies(x, dtype=int)
y = pd.get_dummies(y, dtype=int)
```

3. Visualisasi Imbalance

```python

import matplotlib.pyplot as plt

plt.figure(figsize=(8, 4))
class_counts.plot(kind='bar', color=['yellow', 'blue'])
plt.title('Class Distribution')
plt.xlabel('Class')
plt.ylabel('Count')
plt.xticks(ticks=[0, 1], labels=['Edible', 'Poisonous'], rotation=0)
plt.show()
```

4. Split data test & train

```python
from sklearn.model_selection import train_test_split, GridSearchCV

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

5. Training Decision Tree dengan parameter default

```python
dt = DecisionTreeClassifier()

# set training
dt.fit(X_train, y_train)

# memprediksi label set test
y_test_pred_dt = dt.predict(X_test)

# menghitung akurasi test
acc_test_dt = accuracy_score(y_test, y_test_pred_dt)
print(f"Test accuracy: {acc_test_dt*100:.2f}%")

# Memprediksi label training
y_train_pred_dt = dt.predict(X_train)

#  menghitung akurasi train
acc_train_dt = accuracy_score(y_train, y_train_pred_dt)
print(f"Train accuracy: {acc_train_dt*100:.2f}%")
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree

plt.figure(figsize=(22, 11))
plot_tree(dt, filled=True, feature_names=x.columns.tolist(), class_names=['Edible', 'Posion'])
plt.title("Decision Tree")
plt.show()
```

6. Training Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

# menggunakan estimator atau jumlah pohon,
# saya mencoba dengan nilai 8
rf = RandomForestClassifier(n_estimators=8, random_state=42)

# Sesuaikan data train dan test ke fit
rf.fit(X_train, y_train)

# Prediksi label pred test
y_pred_rf = rf.predict(X_test)

# Hitung Akurasi test
acc_test_rf = accuracy_score(y_test, y_pred_rf)
print(f"Test set accuracy: {acc_test_rf*100:.2f}%")

# Prediksi label pred train
y_pred_train_rf = rf.predict(X_train)

# Hitung Akurasi train
acc_train_rf = accuracy_score(y_train, y_pred_train_rf)
print(f"Train set accuracy: {acc_train_rf *100:.2f}%")
```
